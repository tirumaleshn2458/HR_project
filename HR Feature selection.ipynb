# HR prediction

## Feature selection

```python
import pandas as pd
import numpy as np
```

```python
data_frame=pd.read_csv('modified_data.csv')
```

```python
data_frame.head()
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import Lasso
```

```python
model=SelectFromModel(Lasso(alpha=0.001,random_state=0))
```

```python
x=data_frame.drop(['Attrition'],axis=1)
y=data_frame[['Attrition']]
```

```python
model.fit(x,y)
```

```python
model.get_support()
```

```python
features_to_select=[]
for i,j in zip(x.columns,model.get_support()):
    if j==True:
        features_to_select.append(i)
```

```python
final_data=pd.concat([data_frame[features_to_select],data_frame[['Attrition']]],axis=1)
```

```python
final_data.head()
```

```python
final_data.to_csv('final_data.csv',index=False)
```
